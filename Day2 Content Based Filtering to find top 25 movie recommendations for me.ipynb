## Preprocessing

```python
#removing movie's year from title column and making a new column for it

movies_df['year of release'] = movies_df.title.str.extract("(\(\d\d\d\d\))", expand=False)
movies_df.head(5)
```

```python
#removing parentheses

movies_df['year of release'] = movies_df.title.str.extract("(\d\d\d\d)", expand=False)
movies_df.head(5)
```

```python
#removing year from title column

movies_df['title'] = movies_df.title.str.replace("(\(\d\d\d\d\))", '')
movies_df.head(5)
```

```python
#applying strip function to get rid of extra spaces

movies_df['title'] = movies_df['title'].apply(lambda x: x.strip())
movies_df.head(5)
```

```python
#applying split function to of Genres column

movies_df['genres'] = movies_df.genres.str.split('|')
movies_df.head()
```

# Applying One Hot Encoding to Genres

```python
movies_df
```

```python
movies_genres = movies_df.copy()

#for every row, iterate through list of genres and put 1 for corresponding column
for i,row in movies_df.iterrows():
    for genre in row['genres']:
        movies_genres.at[i,genre]=1

#filling NaN values with 0 which signifies movie does not have that genre
movies_genres= movies_genres.fillna(0)
movies_genres.head(5)
```

```python
#dropping timestamp column from ratings column

ratings_df = ratings_df.drop('timestamp',1)
ratings_df.head(5)
```

# Content Based recommendation system 

```python
#creating a list of my top movies and my ratings for those movies

my_movies = [
    {'title':"Inglourious Basterds",'rating':9.5},
    {'title':"Django Unchained",'rating':8.5},
    {'title':"Jumanji",'rating':8},
    {'title':"Only Lovers Left Alive",'rating':9},
    {'title':"Swades: We, the People (Our Country)",'rating':8.7}
]
```

# Learning my preferences

```python
#filtering out movies 

mansi_Movies = movies_genres[movies_genres['movieId'].isin(input_movies['movieId'].tolist())]
```

```python
mansi_Movies
```

# Cleaning the Data

```python
#resetting the index
mansi_Movies = mansi_Movies.reset_index(drop=True)
mansi_Movies
```

```python
#dropping movieId title genres and year columns becuase we only need the actual genre table

mansi_Genres = mansi_Movies.drop("movieId",1).drop("title",1).drop("genres",1).drop("year of release",1)
```

```python
mansi_Genres
```

# User Profile

We will now turn each genre into weight.
It will be achieved by mutiplying the mansi_Genres table and my ratings from input_movies table and summing up the resulting table by column.

```python
userProfile = mansi_Genres.transpose().dot(input_movies['rating'])

userProfile
```

Let us now extract the genre table from the original df

```python
genreTable = movies_genres.set_index(movies_genres['movieId'])

genreTable = genreTable.drop('movieId',1).drop('title',1).drop('genres',1).drop('year of release',1)
```

```python
genreTable.shape
```

```python
genreTable.head(5)
```

```python
#multiplying genres by weights and then taking the weighted average
s=userProfile.sum()

recommendation_table = ((genreTable*userProfile).sum(axis=1)/s)
recommendation_table.head()
```

```python
#sorting in descending order
recommendation_table = recommendation_table.sort_values(ascending=False)
recommendation_table.head()
```
